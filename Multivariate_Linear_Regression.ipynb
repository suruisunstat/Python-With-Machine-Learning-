```python
import numpy as np
import pandas as pd 
from pandas import DataFrame, Series
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
plt.hist(boston.target,bins=50)
plt.xlabel('Price in $1000s')
plt.ylabel('Number of houses')
```

```python
plt.scatter(boston.data[:,5],boston.target)
plt.ylabel('Price in $1000s')
plt.xlabel('Number of rooms')
```

```python
boston_df = DataFrame(boston.data)
boston_df.columns = boston.feature_names
```

```python
boston_df.head()
```

```python
boston_df['Price']=boston.target
```

```python
boston_df.head()
```

```python
sns.lmplot('RM','Price',boston_df)
```

```python
X = boston_df.RM
Y = boston_df.Price
X.shape
```

```python
X = np.vstack(boston_df.RM)
X.shape
```

```python
X = np.array([[value,1] for value in X])
X
```

```python
m, b = np.linalg.lstsq(X,Y)[0]
```

```python
m
```

```python
b
```

```python
plt.plot(boston_df.RM, boston_df.Price,'o')
x = boston_df.RM
plt.plot(x,m*x+b,'r',label='Best fit line')
```

```python
result = np.linalg.lstsq(X,Y)
total_err = result[1]
rmse = np.sqrt(total_err/len(X))
```

```python
"The root mean square error was %.2f" %rmse
```

```python
rmse[0]
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
```

```python
x_multi = boston_df.drop('Price',1)
```

```python
y_target = boston_df.Price
```

```python
lreg.fit(x_multi,y_target)
```

```python
print(' The estimated intercept coefficient is %.2f ' %lreg.intercept_)
```

```python
print("The number of coefficient is %d " %len(lreg.coef_))
```

```python
coeff_df = DataFrame(boston_df.columns)
coeff_df.columns=['Features']
coeff_df['Coefficient Estimate']= pd.Series(lreg.coef_)
```

```python
coeff_df
```

```python
X_train,X_test,Y_train,Y_test = sklearn.cross_validation.train_test_split(x_multi,y_target)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
lreg = LinearRegression()
```

```python
lreg.fit(X_train,Y_train)
```

```python
pred_train = lreg.predict(X_train)
pred_test = lreg.predict(X_test)
```

```python
print("Fit the model with X_train, and the MSE was %.2f" %np.mean((pred_train-Y_train)**2))
```

```python
print("Fit the model with X_test, and the MSE was %.2f" %np.mean((pred_test-Y_test)**2))
```

```python
train = plt.scatter(pred_train,(Y_train-pred_train),c='b',alpha=0.5)
test = plt.scatter(pred_test,(Y_test-pred_test),c='r',alpha=0.5)
plt.legend((train,test),('training','test'),loc='lower left')
plt.hlines(y=0,xmin=-10,xmax=50)
```
